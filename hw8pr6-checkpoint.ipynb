```python
#
# Example #0: creating random y data... and plotting it
#
```

```python
x_years = list(range(2010,2300))  # The years we're investigating (X-axis/independent)
y_CO2 = [random.uniform(38,8003) for x in x_years]
```

```python
#
# We want to plot!  Let's import some graphing libraries
#
import numpy as np               # Numeric python, numpy
import matplotlib.pyplot as plt  # A plotting library, matplotlib 
```

```python
#
# Function for building a plot
#

def build_co2_plot(x_years, y_CO2):
    """Pass in a list of years and a list of carbon amounts to plot."""
    x_values = np.asarray(x_years)  # convert from Python list to numpy array
    y_values = np.asarray(y_CO2)    # ditto

    plt.subplot(1, 1, 1)
    plt.plot(x_values, 
             y_values, 
             '-',          # Connect via line (try "*" or "." or "-")
             lw    = 1,    # Line weight  (so many more options available!)
             # Reference:  matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
             color = "xkcd:teal")   # Yes, really! [[https://xkcd.com/color/rgb/]]   

    plt.xlabel('years')
    plt.ylabel('CO2 (units)')        # Should add units here...
    plt.title('Plotting CO2 vs year')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
    
# build_co2_plot(x_years, y_CO2)
```

```python
#
# Stand-alone plotting example, using the above function
#
build_co2_plot(x_years, y_CO2)
```

```python
#
# Example #1:  computing the data based on a known relationship
#

# Here is the known relationship:
# CO2 in year 2010 was at the value 9001.58 gt
starting_CO2 = 9001.58

# the rate of carbon emissions is 10 gt/year
yearly_delta = 10.0

# we loop to compute the carbon each year!
y_CO2 = [starting_CO2]                         # Starting value!
for year_index in range(1, len(x_years)):      # Already have the starting value!
    year = x_years[year_index]                 # Let's get the "current" year

    old_CO2 = y_CO2[year_index - 1]            # Last year's CO2
    new_CO2 = old_CO2 + yearly_delta           # Compute this year's CO2
    
    y_CO2 = y_CO2 + [new_CO2]                  # Add it to our y_CO2 list
    
    print("year", year, "adding", yearly_delta, "   new total:", new_CO2)
    
print("\n\n")
print("The final list, y_CO2 has", len(y_CO2), "estimates")
```

```python
#
# Stand-alone plotting example, using the above function
#
build_co2_plot(x_years, y_CO2)
```

```python
#
# Example #2:  computing the data based on a known 2nd-derivative relationship
#

# Here is the known relationship:
# CO2 in year 2010 was at the value 9001.58 gt
starting_CO2 = 9001.58

# The rate of carbon emissions is 10 gt/year
yearly_delta = 10.0

# Each year, the rate of carbon emissions is increasing by 1.58 gt/year
co2_acceleration = 1.58

# we loop to compute the carbon each year:
y_CO2 = [starting_CO2]                         # Starting value!
for year_index in range(1, len(x_years)):   # Already have the starting value!
    year = x_years[year_index]                 # Let's get the "current" year
    old_CO2 = y_CO2[year_index - 1]            # And get last year's CO2
    
    # This year's CO2 rate of change is in yearly_delta
    new_CO2 = old_CO2 + yearly_delta           # Compute this year's CO2
    y_CO2 = y_CO2 + [new_CO2]                  # Add this year's to our y_CO2 list
    
    print("year", year, "adding", yearly_delta, "   new total:", new_CO2)
    
    yearly_delta = yearly_delta + co2_acceleration  # Now, update to next year's delta
    
print("\n\n")
print("The final list, y_CO2 has", len(y_CO2), "estimates")
```

```python
#
# Stand-alone plotting example, using the above function
#
build_co2_plot(x_years, y_CO2)
```
